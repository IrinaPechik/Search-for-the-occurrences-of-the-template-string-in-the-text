# Построим графики

```python
def plot_table(file_path, title):
    df = pd.read_csv(file_path, index_col=0) # загрузка таблицы из файла, указываем первый столбец как индекс
    df = df.iloc[:, :-1]
    df.plot(kind='line') # построение графика
    plt.xlabel('size of pattern') # подпись оси x
    plt.ylabel('time, nanoseconds') # подпись оси y
    plt.title(title) # заголовок графика
    plt.legend(loc='upper left') # легенда, где находятся линии на графике
    plt.show() # показать график
```

```python
plot_table("short_bin_text.csv", "Без символов подстановки, 10 тыс. символов, бинарный алфавит")
```

Выводы:
1. **KMP** - работает стабильно, скачки небольшие.
2. **KMP с уточнёнными гранями** работает в основном быстрее, чем KMP. Разница довольно ощутима. Однако, в начале его работы произошёл немаленький скачок.
3. **Алгоритм банального посимвольного сравнения** сначала работает хуже всех, однако потом показывает самый лучший результат. Но в конце происходит огромный скачок.
4. **Алгоритм Рабина-Карпа** почти всегда работает быстрее всех

```python
plot_table("long_bin_text.csv", "Без символов подстановки, 100 тыс. символов, бинарный алфавит")
```

Выводы:
Все алгоритмы имеют одинаковую тенденцию изменения времени работы. Скачки довольно большие и происходят примерно в один момент.
